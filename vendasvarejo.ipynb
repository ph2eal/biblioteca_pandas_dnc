```python
import pandas as pd

import seaborn as sns
```

```python
vendas = pd.read_excel("/content/varejo.xlsx")
vendas.head()
```

```python
#verificar as categorias dentro de uma coluna. ex: categorias do canal venda:
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP','Aplicativo')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby('bandeira').idcompra.nunique()
```

```python
vendas.groupby('estado').idcompra.nunique()
```

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ','_')
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace('Bebes','Bebês')
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.info()
```

```python
vendas[vendas['Preço'].isnull()]
```

```python
vendas[vendas['estado'].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
vendas['estado'].fillna("MS", inplace=True)
vendas.info()
```

```python
media_preco = vendas['Preço'].mean()
media_preco

vendas['Preço'].fillna(media_preco, inplace=True)
vendas.info()
```

```python
#filtrando os dados
vendas[["Preço","Preço_com_frete"]].describe()
```

```python
#tirando os dados onde preço > preço_com_frete

vendas_correto = vendas.query("Preço < Preço_com_frete")
vendas_correto.describe()
```

```python
vendas_correto.info()
```

```python
#iniciando os filtros -ex: lojas departamento Esporte e lazer em SP

vendas_correto.query ("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'" ).head(50)
```

```python
#iniciando os filtros -ex: id canal de venda MOBILE

vendas_correto.query ("idcanalvenda == 'Mobile' and estado == 'SP'").describe()
```

```python
# INDICADOR 1: TOP 5 - DEPARTAMENTOS MAIS VENDIDOS

vendas_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index().head()
```

```python
# INDICADOR 2: TOP 5 - DEPARTAMENTOS MENOS VENDIDOS

vendas_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=True).reset_index().head()
```

```python
# INDICADOR 3: TOP 5 -  MÉDIA DE PREÇO POR NOME DE DEPARTAMENTO

round(vendas_correto.groupby('Nome_Departamento')["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index().head(),2)
```

```python
# INDICADOR 4: QUANTIDADE DE VENDAS POR MÊS
#4.1. POR DIA
vendas_correto.groupby('Data').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
# INDICADOR 4: QUANTIDADE DE VENDAS POR MÊS
#4.1. POR MÊS.
#4.1.1. VERIFICAR O MÊS

vendas_correto["mes"] = vendas_correto["Data"].dt.month
vendas_correto.head()
```

```python
# INDICADOR 4: QUANTIDADE DE VENDAS POR MÊS
#4.1. POR MÊS.
#4.1.2. AGRUPAR POR MÊS


vendas_correto.groupby('mes').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
#IMPORTANDO NOVA BASE DE DADOS (CLIENTES)
clientes = pd.read_excel("cliente_varejo.xlsx")
clientes.head()
```

```python
clientes[["idade","renda"]].describe()
```

```python
clientes.info()
```

```python
#mudando a renda de INT pra FLOAT
clientes = clientes.astype({"renda": float})
clientes.info()
```

```python
vendas_correto.head()
```

```python
vendas_cliente = vendas_correto.merge(clientes, how = "left", on = "cliente_Log")
vendas_cliente.head()
```

```python
vendas_cliente.info()
```

```python
vendas_cliente.head()
```

```python
vendas_cliente = vendas_cliente.rename(columns={"renda_y": "renda"})
vendas_cliente = vendas_cliente.rename(columns={"idade_y": "idade"})
vendas_cliente.head()
```

```python
# 5 - MÉDIA DE RENDA PARA CADA CANAL DE VENDA
#5.1. AGRUPAMENTO ID CANAL RENDA
agg_canal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_canal_renda
```

```python
#6 - MÉDIA DE IDADE DOS CLIENTES POR BANDEIRA
#6.1. AGRUPAMENTO BANDEIRA IDADE
agg_bandeira_idade = round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_bandeira_idade
```

```python
# VISUALIZAÇÕES
import matplotlib.pyplot as plt
```

```python
#GRÁFICO MÉDIA DE IDADE POR BANDEIRA

fig = plt.figure (figsize= (7,5))
plt.bar (agg_bandeira_idade['bandeira'], agg_bandeira_idade['idade'], color = "green");
plt.title ("Média de idade dos clientes por bandeira", loc = "center", fontsize = 15, fontweight = "bold", color = "black");
plt.ylabel ("Média de Idade")
plt.xlabel ("Bandeira")
```

```python
#GRÁFICO - MÉDIA DE IDADE POR CANAL DE VENDA

fig = plt.figure (figsize= (7,5))
plt.bar(agg_canal_renda['idcanalvenda'], agg_canal_renda['renda'], color = "blue");
plt.title ("Média de renda por canal de venda", loc = "center", fontsize = 15, fontweight = "bold", color = "black");
plt.ylabel ("Média de Renda")
plt.xlabel ("Canal de Venda")
```

```python
#QUANTIDADE DE VENDAS POR DATA:

vendas_por_data = vendas_correto.groupby('Data').idcompra.nunique().reset_index()
vendas_por_data.head()
```

```python
# GRÁFICO DE LINHAS VENDAS POR DATA:

fig = plt.figure (figsize= (10,20))
plt.xlabel ("Data de venda")
plt.ylabel ("Quantidade de Vendas")
x = vendas_por_data['Data']
y = vendas_por_data['idcompra']
plt.plot (x,y, color = "red", linewidth = "3")
plt.title ("Quantidade de vendas por data", loc = "center", fontsize = 15, fontweight = "bold", color = "black");
plt.show()
```

```python
#GRÁFICO DE MÉDIA DE PREÇO COM FRETE POR NOME DE DEPARTAMENTO


agg_dept_preco = round(vendas_correto.groupby('Nome_Departamento')["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index().head(),2)
agg_dept_preco


```

```python
fig = plt.figure (figsize= (15,15))
plt.bar(agg_dept_preco['Nome_Departamento'], agg_dept_preco['Preço_com_frete'], color = "yellow");

plt.ylabel ("Média de preço com frete")
plt.xlabel ("Nome de departamento")

plt.title ("Média de preço com frete por nome de departamento", loc = "center", fontsize = 15, fontweight = "bold", color = "black");
```
